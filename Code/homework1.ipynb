```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
```

```python
family=nx.florentine_families_graph();
nx.draw_networkx(family, node_size=30, alpha=0.5, node_color="blue", font_size='12',with_labels=True)
plt.show()
```

```python
if nx.connected.is_connected(family) == True: 
    print("The network is full connected.")
else:
    print("The network is not fully connected.")
```

```python
print(list(nx.articulation_points(family)))
for node in nx.articulation_points(family): 
    family_temp = family.copy() 
    family_temp.remove_node(node) 
    print('Number of connected components after deleting {}:{}'.format(node,str(nx.number_connected_components(family_temp))))
    nx.draw_networkx(family_temp, node_size=30, alpha=0.5, node_color="blue", font_size='12',with_labels=True)
    plt.show()
```

Removal of 'Medici' family disconnects the graph into 3 connected components. Comparing to 2 connected components caused by the removal of 'Guadagni', 'Medici', 'Salviati', 'Medici' family is therefore more remarkable.
